# Pandas DataFrame

```python
import pandas as pd
import numpy as np
```

## Creating DataFrame

```python
# Data
a= {"Name":[chr(i) for i in range(97,107)],
    "Age":[i for  i in range(20,30)],
    "City":[chr(i) for i in range(65,75)]
}
```

```python
a
```

```python
#Creating Data Frame
df=pd.DataFrame(a,index=[i for i in range(100,110)])      # By using Dictionary keys are assigned as Column names 
df
```

```python
# If DF is created by this column names are given using its parameters
pd.DataFrame(np.array([[1,2,3],[4,5,6]]),columns=['a','b','c'])
```

## Data Cleaning / Manipulation

```python
df1= pd.DataFrame({'Name': ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'a','c'],
 'Age': [20, 21, 22, 23, None, None, 26, 27, None, 20,22],
 'City': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'A','C']})
df1
```

```python
# Drop lables 
df1.drop([2,3])     # Drop Specified rows or columns
```

```python
# Drop Duplicates
df1.drop_duplicates()       # To make it permanent we have to give inplace= True in Parameter
```

```python
# Fill Null Values
df1.fillna(df1.Age.mean())  # Can be filled with mean , median etc
```

```python
# Replace 
df1.replace('A','Changes occur Here')
```

```python
#Rename 
df1.rename(index={0:"x",1:"y",2:"z"}, columns={"Name":"Lowercase","Age":"Numbers","City":"Upper"})
```

```python
# Sort_values 
data = {
    'City': ['New York', 'Los Angeles', 'Chicago', 'New York', 'Chicago',
             'Los Angeles', 'New York', 'Chicago', 'Los Angeles', 'New York'],
    'Product': ['Apple', 'Orange', 'Apple', 'Banana', 'Orange',
                'Banana', 'Apple', 'Banana', 'Orange', 'Apple'],
    'Sales': [100, 150, 200, 120, 180, 90, 300, 130, 160, 250]
}
df2=pd.DataFrame(data)
df2
```

```python
df2.sort_values(by="Sales",ascending=False)
```

```python
df2.groupby(by='City')['Sales'].aggregate({'mean','median','min','max'})
```

```python
df2.groupby(by='City')['Sales'].aggregate({'mean','median','min','max'}).reset_index()
```

```python
c=df2.pivot_table(index='City',columns='Product',values='Sales',aggfunc='sum')
c
```

```python
df2.melt(id_vars=['City'],value_vars=['Product'])
```

## Arithmetic and Statistical Operations

```python
print(df2.Sales.mean())        # Aggregate function 
print(df2.Sales.median())        
print(df2.Sales.sum())
print(df2.Sales.min())
print(df2.Sales.max())
```

```python
# Statistical Funtions
print(df2.Sales.std())
print(df2.Sales.var()) 
```

```python
df3=pd.DataFrame(data = {
    'X': [2.5, 3.1, 4.0, 5.7, 6.8, 8.2, 7.1, 9.6, 10.5, 12.2,
          13.3, 14.8, 15.0, 16.4, 17.2, 18.5, 19.1, 20.7, 21.5, 22.3],
    'Y': [3.0, 4.2, 6.5, 5.8, 12.0, 11.5, 14.0, 19.1, 18.2, 20.5,
          22.0, 24.5, 29.0, 30.5, 27.2, 34.0, 35.5, 36.2, 39.0, 42.5]
})
```

```python
df3.pct_change().head(7) # Fractional change between the current and a prior element.
```

## Combining DataFrame

```python
pd.concat((df2,df2),axis=1)
```

```python
j1=pd.DataFrame(data = {
    'EmployeeID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'DepartmentID': [101, 102, 101, 103, 102]
})
j2=pd.DataFrame(data = {
    'DepartmentID': [101, 102, 103, 104],
    'DepartmentName': ['HR', 'IT', 'Finance', 'Marketing']
})
```

```python
# Simple Merge
pd.merge(j1,j2,how='inner',on='DepartmentID')
```

```python
j1.set_index('DepartmentID',inplace=True)
j2.set_index('DepartmentID',inplace=True)
```

```python
# Simple Join
j1.join(j2)
```

## Reshaping Data

```python
r=pd.DataFrame(data = {
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04'],
    'Sales': [200, 220, 250, 275],
    'Expenses': [150, 160, 170, 180]
})
r.set_index('Date',inplace=True)
r
```

```python
# Basic Stacking
r.stack(level=0)
```

```python
r=pd.DataFrame({
    'Product': ['A', 'A', 'B', 'B'],
    'Month': ['Jan', 'Feb', 'Jan', 'Feb'],
    'Sales': [100, 150, 200, 250]
})
r.set_index(['Product', 'Month'], inplace=True)
r
```

```python
# Multi Index Stack 
r.stack()
```

```python
# Transpose
df1.transpose()
```

```python
df1.T       # .T  also does the same
```

```python
# Resets the Index
r.reset_index(inplace=True)
r
```

```python
# Set Index
r.set_index('Product')
```
